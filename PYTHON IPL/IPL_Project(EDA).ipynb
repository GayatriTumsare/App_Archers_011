Import all the required Librabries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import zipfile
import os
```

Check all the Files in the zip_Folder and extract it

```python
# Path to the uploaded ZIP file
zip_file_path = "C:/Users/ASUS/Desktop/Project IPL/IPL_2016.zip"

# Extract the ZIP file
extracted_dir = "C:/Users/ASUS/Desktop/Project IPL/IPL_2016_cleaned"
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_dir)

# List the extracted files
extracted_files = os.listdir(extracted_dir)
extracted_files
```

Load the CSV files into pandas DataFrames

```python
# Removing the problematic feature engineering part for 'Batsman_Scored'
# Proceeding with the rest of the EDA and data processing

# Load the CSV files into pandas DataFrames
dataframes = {}
for file in extracted_files:
    if file.endswith('.csv'):
        df_name = os.path.splitext(file)[0]
        file_path = os.path.join(extracted_dir, file)
        try:
            dataframes[df_name] = pd.read_csv(file_path)
        except:
            dataframes[df_name] = pd.read_csv(file_path,error_bad_lines=False)
```

Display basic information and first few rows for each DataFrame

```python
for name, df in dataframes.items():
    print(f"\n--- {name} ---")
    print(df.info())
    print(df.head())
```

Handle missing values by identifying and dropping or imputing

```python
for name, df in dataframes.items():
    print(f"\nMissing values in {name}:")
    print(df.isnull().sum())
    # Drop rows with missing values (example approach)
    df.dropna(inplace=True)
    dataframes[name] = df
```

Check and remove duplicates

```python
for name, df in dataframes.items():
    print(f"\nDuplicates in {name}: {df.duplicated().sum()}")
    df.drop_duplicates(inplace=True)
    dataframes[name] = df
```

Save the cleaned files in csv format

```python
# Directory where the cleaned CSV files will be saved
cleaned_dir = "C:/Users/ASUS/Desktop/Project IPL/IPL_2016/IPL_clean_files"
# cleaned_dir = '/Users/saswatranjannayak/Desktop/ASE204 Project/IPL_2016/clean'

# Create the directory if it doesn't exist
os.makedirs(cleaned_dir, exist_ok=True)

# Save each cleaned DataFrame to a CSV file
for name, df in dataframes.items():
    cleaned_file_path = os.path.join(cleaned_dir, f'IPL_{name}.csv')
    df.to_csv(cleaned_file_path, index=False)
    print(f'Saved {name} to {cleaned_file_path}')
```
